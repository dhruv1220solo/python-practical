Program to display RF bands using if–elif

```python
a = float(input("Enter frequency in Hz: "))
if 30000 <= a <= 300000:
    print("VLF Band")
    
elif 300000 <=a <=3000000:
    print("LF Band")

elif 300000 <= a <=3000000:
    print ("HF Band")

elif 3000000 <= a <= 30000000:
    print ("VHF Band")

else:
    print("Frequency not in standard RF range")
```

Write a Python program that checks if a resistor value is within a tolerance range (e.g., 5% tolerance for a 100Ω resistor) and prints whether it is within tolerance or not.

```python
resistor_value = float(input("Enter the measured resistor value (in ohms): "))

nominal = 100
tolerance = 0.05 * nominal

lower_limit = nominal - tolerance
upper_limit = nominal + tolerance

if lower_limit <= resistor_value <= upper_limit:
    print("Resistor is within tolerance")
else:
    print("Resistor is out of tolerance")
```
